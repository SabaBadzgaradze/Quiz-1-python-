```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ჩატვირთეთ CSV ფაილი DataFrame-ში
df = pd.read_csv('survey_results_public.csv')
df.head()
```

```python
# ამოიღეთ ცხრილის სასურველი სტრიქონები და სვეტები
desired_data = df[['MainBranch', 'Age', 'Employment', 'YearsCode', 'YearsCodePro', 'Country', 'ConvertedCompYearly']]
desired_data.head()
```

```python
# მიანიჭეთ ინდექსირება კონკრეტულ სვეტს
desired_data.set_index('MainBranch', inplace=True)
```

```python
# შექმენით ფილტრი 2 პარამეტრზე დაყრდნობით
filtered_data = df[(df['Country'] == 'United States of America') & (df['Employment'] == 'Employed, full-time')]
filtered_data.head()
```

```python
# დაასორტირეთ ცხრილი 2 პარამეტრის გამოყენებით
sorted_data = df.sort_values(by=['Age', 'ConvertedCompYearly'])
sorted_data.head()
```

```python
# სტატისტიკური ფუნქციები
print("\nსტატისტიკური რეზიუმე:")
print("კონვერტირებული კომპენსაციის საშუალო (წლიური):", df['ConvertedCompYearly'].mean())
print("კონვერტირებული კომპენსაციის სტანდარტული გადახრა (წლიურად):", df['ConvertedCompYearly'].std())
print("კონვერტირებული კომპენსაციის მედიანა (წლიურად):", df['ConvertedCompYearly'].median())
print("მინიმალური გადაყვანილი კომპენსაცია (წლიურად):", df['ConvertedCompYearly'].min())
print("მაქსიმალური კონვერტაციის კომპენსაცია (წლიურად):", df['ConvertedCompYearly'].max())
```

```python
# დეველოპერების რაოდენობა დასაქმების სტატუსის მიხედვით
employment_counts = df['Employment'].value_counts().head(5)
employment_counts.plot(kind='bar', color='skyblue')
plt.title('დეველოპერების დასაქმების სტატუსი')
plt.xlabel('დასაქმების სტატუსი')
plt.ylabel('Count')
plt.show()
```

```python
# Building Line Chart
age_salary_data = df.groupby('Age')['ConvertedCompYearly'].median().reset_index().head(5)
age_salary_data.plot(x='Age', y='ConvertedCompYearly', marker='o', color='orange')
plt.title('საშუალო ხელფასი ასაკის მიხედვით')
plt.xlabel('ასაკი')
plt.ylabel('საშუალო წლიური ანაზღაურება')
plt.grid(True)
plt.show()
```
